```python
def getPermutation( n, k):
    return helper([x + 1 for x in range(n)], n, k)

def helper(nums, n, k):
    
    print(nums, n, k)
    if n == 1:
        return str(nums[0])

    i = (k - 1) // math.factorial(n - 1)
    return str(nums[i]) + helper(nums[:i] + nums[i + 1:], n - 1, (k - 1) % math.factorial(n - 1) + 1)
```

```python
def permutations(s):
    if len(s) == 1:
        return s

    recursive_perms = set()
    for c in s:
        for perm in permutations(s.replace(c,'',1)):
            recursive_perms.add(c+perm)

    return set(recursive_perms)
```
